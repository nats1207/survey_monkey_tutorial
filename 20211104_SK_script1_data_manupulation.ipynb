```python
# import libraries
import pandas as pd
import os
```

```python
# define current working directory
cwd = os.getcwd()
```

```python
# import dataset
dataset = pd.read_excel(cwd + "/Data - Survey Monkey Output - Edited.xlsx", sheet_name="edited_data")
dataset.head()
```

```python
# make a copy of the dataset
dataset_modified = dataset.copy()
dataset_modified.head()
```

```python
# take a look at all the columns and drop the ones we do not need
dataset_modified.columns
columns_to_drop = ['Start Date', 'End Date', 'Email Address', 'First Name', 'Last Name', 'Custom Data 1']
dataset_modified = dataset_modified.drop(columns=columns_to_drop, errors='ignore')
dataset_modified.head()
```

```python
# unpivot the data

# define variables for melt function
id_vars = list(dataset_modified.columns[:8])
value_vars = list(dataset_modified.columns[8:])
var_name = 'Question - subquestion'
value_name = 'Answer'

# melt
dataset_melted = dataset_modified.melt(id_vars=id_vars, value_vars=value_vars, var_name=var_name, value_name=value_name)
dataset_melted.sort_values('Respondent ID', ascending=True, inplace=True)
dataset_melted
```

```python
# add original question using split
dataset_melted['Question'] = dataset_melted['Question - subquestion'].str.split(' - ').str[0]
dataset_melted
```
